```python
datamat = dataset[dataset.columns[[5,6,4]]].values
datamat = datamat[~np.isnan(datamat).any(axis=1)]
X_train = datamat[:,0:2]
y_train = datamat[:,2]
```

Most of these graphs shows positive linear correlations. The factors that shows this correlation will be the features used for a regression method.

```python
# Transforming useful features into log-scale
dataset['Max case'] = dataset['Max case'].apply(np.log)
dataset['densityMi'] = dataset['densityMi'].apply(np.log)
dataset['Flights'] = dataset['Flights'].apply(np.log)
dataset['Tourists'] = dataset['Tourists'].apply(np.log)

processed_dataset = dataset.replace([np.inf,-np.inf],0)
processed_dataset
```

I will apply Nueral Networks to curve-fit the feature space. Using the classical Linear Regression method is fine too. The following are the hyper-parameters for the Neural Network:

* Loss function: MSE
* Optimizer: Adam
* Weight Initializers of N(0,0.01^2)
* Learning rate: 0.001
* Number of epochs = 100
* Validation split: .45

```python
import keras
from keras import initializers
from keras.models import Sequential
from keras.layers import Dense, Activation
from keras.optimizers import Adam
import tensorflow as tf

ndataset = (processed_dataset-processed_dataset.mean())/processed_dataset.std()
training = ndataset[ndataset.columns[[2,5,6,4]]].values
training = training[~np.isnan(datamat).any(axis=1)]
X_train = training[:,0:3]
y_train = training[:,3]
epochs = 100

model = Sequential()
model.add(Dense(32,activation = 'relu', input_shape=(3,),kernel_initializer= initializers.RandomNormal(stddev=0.01)))
model.add(Dense(8,activation = 'relu'))
model.add(Dense(1,activation = 'linear'))
model.compile(loss = 'mse', optimizer = Adam(lr=0.001))

history = model.fit(X_train,y_train,validation_split=0.45, batch_size=5, epochs = epochs, verbose= 0,shuffle = True)
w = model.layers[0].get_weights()
loss_values = history.history['loss']
valloss_values = history.history['val_loss']
epochs = range(1, len(loss_values)+1)

plt.plot(epochs, loss_values, label='Training Loss')
plt.plot(epochs, valloss_values, label='Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

Ideally, the output of a neural network is expressed as 
$$ \hat{\textbf{y}} =\textbf{x}^T \textbf{w}$$

The cost function is the Mean-Squared error, which is expressed as follows.
$$L(x,y) = \frac{1}{n} (\textbf{y} - \hat{\textbf{y}})^T (\textbf{y} - \hat{\textbf{y}})$$
where $\textbf{y} \in \mathbb{R}^n$ is the the target vector and $\hat{\textbf{y}} \in \mathbb{R}^n$ is the estimated target vector.
The general goal of the optimization algorithm is to find $\hat{w}$ at which
$$\hat{w} = argmin_w L(x,y)$$
The trained weight matrix serves as the regression line in the feature space. Using this, we can project the max case by using the features selected.
